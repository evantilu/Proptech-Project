## Part I
#### Inspecting datasets

```python
import pandas as pd
```

```python
irv_urb = pd.read_csv('data/irvine_ca_city_2.csv')
irv_sub = pd.read_csv('data/irvine_ca_suburb_2.csv')
san_diego_urb = pd.read_csv('data/san_diego_ca_city_2.csv')
san_diego_sub = pd.read_csv('data/san_diego_ca_suburb_2.csv')
san_fran_urb = pd.read_csv('data/san_francisco_ca_city_2.csv')
san_fran_sub = pd.read_csv('data/san_francisco_ca_suburb_2.csv')
```

### Check empty fields for each col in every dataset

```python
tmp_1 = irv_urb.isnull().any()
print('irv_urb fields with empty values')
print('-------------------------------')
print(tmp_1[tmp_1])
print()

tmp_2 = irv_sub.isnull().any()
print('-------------------------------')
print('irv_sub fields with empty values')
print(tmp_2[tmp_2])
print()

tmp = san_diego_urb.isnull().any()
print('-------------------------------')
print('san_diego_urb fields with empty values')
print(tmp[tmp])
print()

tmp = san_diego_sub.isnull().any()
print('-------------------------------')
print('san_diego_sub fields with empty values')
print(tmp[tmp])
print()

tmp = san_fran_urb.isnull().any()
print('-------------------------------')
print('san_fran_urb fields with empty values')
print(tmp[tmp])
print()

tmp = san_fran_sub.isnull().any()
print('-------------------------------')
print('san_fran_sub fields with empty values')
print(tmp[tmp])
print()
```

From above, there's not much info so after manual inspection, I confirmed that all datasets have the 
same fields that have empty values. And none of the fields are the critical ones that must not be dropped.
